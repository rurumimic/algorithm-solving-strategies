```python
# 1은 무조건 맨 오른쪽에
# 2는 첫줄꽉 중간 꽉 마지막 꽉
# 나머지는 한쪽
# 3일때 5
# 5일때 9=> 2*N - 1
# 7일때 13?

# Input 개수

    
def t_x(result, sort, size, max_size):
    Y = 0
    if sort == "TOP":
        Y = 0
        pass
    elif sort == "MIDDLE":
        Y = max_size - size
        pass
    elif sort == "BOTTOM":
        Y = max_size - 1
        pass
    
    if size == max_size:
        Y = 0
        pass

    for i in range(size):
        result[Y][i] = "#"
        pass

def m_x(result, sort, size, max_size):
    Y = 0
    if sort == "TOP":
        Y = size - 1
        pass
    elif sort == "MIDDLE":
        Y = max_size - 1
        pass
    elif sort == "BOTTOM":
        Y = 2 * max_size - size - 1
        pass
    
    if size == max_size:
        Y = max_size - 1
        pass

    for i in range(size):
        result[Y][i] = "#"
        pass

def b_x(result, sort, size, max_size):
    Y = 0
    if sort == "TOP":
        Y = max_size - 1
        pass
    elif sort == "MIDDLE":
        Y = 2 * max_size - size - 1
        pass
    elif sort == "BOTTOM":
        Y = 2 * max_size - 2
        pass
    if size == max_size:
        Y = 2 * max_size - 2
        pass

    for i in range(size):
        result[Y][i] = "#"
        pass

def lt_y(result, sort, size, max_size):
    X = 0
    if sort == "TOP":
        X = 0
        pass
    elif sort == "MIDDLE":
        X = max_size - size
        pass
    elif sort == "BOTTOM":
        X = max_size - 1
        pass
    if size == max_size:
        X = 0
        pass
    for i in range(X, X + size):
        result[i][0] = "#"
        pass

def rt_y(result, sort, size, max_size):
    X = 0
    if sort == "TOP":
        X = 0
        pass
    elif sort == "MIDDLE":
        X = max_size - size
        pass
    elif sort == "BOTTOM":
        X = max_size - 1
        pass
    if size == max_size:
        X = 0
        pass
    for i in range(X, X + size):
        result[i][size - 1] = "#"
        pass

def lb_y(result, sort, size, max_size):
    X = 0
    if sort == "TOP":
        X = max_size - size
        pass
    elif sort == "MIDDLE":
        X = max_size - 1
        pass
    elif sort == "BOTTOM":
        X = 2 * max_size - size - 1
        pass
    if size == max_size:
            X = max_size - 1
            pass
    for i in range(X, X + size):
        result[i][0] = "#"
        pass

def rb_y(result, sort, size, max_size):
    X = 0
    if sort == "TOP":
        X = max_size - size
        pass
    elif sort == "MIDDLE":
        X = max_size - 1
        pass
    elif sort == "BOTTOM":
        X = 2 * max_size - size - 1
        pass
    if size == max_size:
        X = max_size - 1
    for i in range(X, X + size):
        result[i][size - 1] = "#"
        pass      

def make_num(number, sort, X, Y):
    # x에는 size
    # y에는 max_size
    result = []
        
    for y in range((2 * Y) - 1):
        number_x = []
        for x in range(X):
            number_x.append(".")
            pass
        result.append(number_x)
        pass
    pass

    if number == "0":
        t_x(result, sort, X, Y)
        b_x(result, sort, X, Y)
        lt_y(result, sort, X, Y)
        lb_y(result, sort, X, Y)
        rt_y(result, sort, X, Y)
        rb_y(result, sort, X, Y)
        pass
    elif number == "1":
        rt_y(result, sort, X, Y)
        rb_y(result, sort, X, Y)
        pass
    elif number == "2":
        t_x(result, sort, X, Y)
        m_x(result, sort, X, Y)
        b_x(result, sort, X, Y)
        lb_y(result, sort, X, Y)
        rt_y(result, sort, X, Y)
        pass
    elif number == "3":
        t_x(result, sort, X, Y)
        m_x(result, sort, X, Y)
        b_x(result, sort, X, Y)
        rt_y(result, sort, X, Y)
        rb_y(result, sort, X, Y)
        pass
    elif number == "4":
        m_x(result, sort, X, Y)
        lt_y(result, sort, X, Y)
        rt_y(result, sort, X, Y)
        rb_y(result, sort, X, Y)
        pass
    elif number == "5":
        t_x(result, sort, X, Y)
        m_x(result, sort, X, Y)
        b_x(result, sort, X, Y)
        lt_y(result, sort, X, Y)
        rb_y(result, sort, X, Y)
        pass
    elif number == "6":
        m_x(result, sort, X, Y)
        b_x(result, sort, X, Y)
        lt_y(result, sort, X, Y)
        lb_y(result, sort, X, Y)
        rb_y(result, sort, X, Y)
        pass
    elif number == "7":
        t_x(result, sort, X, Y)
        rt_y(result, sort, X, Y)
        rb_y(result, sort, X, Y)
        pass
    elif number == "8":
        t_x(result, sort, X, Y)
        m_x(result, sort, X, Y)
        b_x(result, sort, X, Y)
        lt_y(result, sort, X, Y)
        lb_y(result, sort, X, Y)
        rt_y(result, sort, X, Y)
        rb_y(result, sort, X, Y)
        pass
    elif number == "9":
        t_x(result, sort, X, Y)
        m_x(result, sort, X, Y)
        lt_y(result, sort, X, Y)
        rt_y(result, sort, X, Y)
        rb_y(result, sort, X, Y)
        pass
    
    return result

if __name__=="__main__":
    lists = input().split(" ")
    N = int(lists[0])
    sort = lists[1]

    sizes = []
    numbers = []

    for _ in range(N):
        lists = input().split(" ")
        sizes.append(int(lists[0]))
        numbers.append(lists[1])
        pass

    # print(sizes)
    # print(numbers)

    result = []
    max_size = max(sizes)

    count = 0
    for i in range(N):
        X = sizes[i]
        Y = max_size
        for number in numbers[i]:
            result.append(make_num(number, sort,X , Y))
            count+=1
            pass
        pass

    # for num in result:
    #     for stk in num:
    #         print(stk)

    for i in range(2 * max_size - 1):
        for j in range(count):
            for answer in result[j][i]:
                print(answer, end="")
            if j == count - 1:    
                print("")
            else:
                print(" ", end="")
```
